## Least Squares

Least squares is a fundamental statistical technique used for parameter estimation in regression analysis. Its primary objective is to minimize the sum of the squares of the deviations (errors) between the observed values and the predicted values generated by the model.
####Definition

The least squares method seeks to find the best-fitting line through a set of data points by minimizing the residual sum of squares (RSS). The residuals are the differences between the observed values and the values predicted by the model. Mathematically, this can be expressed as:

image.png


Iterative Process: The method can be thought of as an iterative search algorithm that explores different combinations of slopes and intercepts. It does so by:

- Testing Different Parameters: The algorithm begins with initial guesses for the slope and intercept.

- Calculating Predictions: For each combination of slope and intercept, it calculates the predicted values.

- Evaluating Deviations: It measures the deviations (or errors) of these predictions from the actual observed values.

- Adjusting Parameters: Based on the sum of squared deviations, the algorithm adjusts the parameters slightly to minimize the RSS. This step continues iteratively until the changes in parameters no longer result in significant reductions in the RSS.


This is very important when discussing logistic reggression!



##Least squares implementation of Linear reggression

In practice, we use least squares regression to minimize the difference between the actual data points and the predicted values, which gives us the best-fitting line.

---

The Least Squares Method calculates the slope `m` and intercept `b` by minimizing the sum of the squared difference between the observed values `y` and the predicted values.

Mathematically, the line predicted is simply:

```
y = mx + b
```
where

```
b = y - mx
```

- y  --> mean y values
- x --> mean  x values
- m is the slope of the line of which is calculated using the formula above

# Logistic Regression


Logistic regression is a statistical method used for binary classification, predicting the probability that an outcome belongs to one of two categories (e.g., success/failure). Unlike linear regression, which predicts continuous values, logistic regression uses the logistic function (or sigmoid function) to map predicted values to probabilities between 0 and 1. The model estimates coefficients through maximum likelihood estimation, allowing it to find the best-fitting line in a log-odds space. Commonly applied in fields like medicine, social sciences, and marketing, logistic regression provides interpretable results and a probabilistic framework for decision-making, assuming independence among observations and a linear relationship between independent variables and the log odds of the dependent variable.

##Gradient Descent

Gradient descent is an optimization algorithm commonly used in machine learning and statistics to minimize a cost function by iteratively adjusting model parameters. The goal is to find the parameter values that minimize the difference between the predicted output and the actual output.


> It can be prone to overfitting, that is why iterations are important



---



---


1. Gradient descent starts with an initial guess for the model parameters (weights). These can be set randomly or initialized to zero.



2. At each iteration, the algorithm computes the gradient of the cost function with respect to the model parameters. The gradient is a vector of partial derivatives that indicates the direction of steepest ascent of the function.

3. The parameters are then updated by moving in the opposite direction of the gradient. This is because the goal is to minimize the cost function:

4. Steps 3 and 4 are repeated until the algorithm converges, which occurs when changes to the cost function become negligible or a predefined number of iterations is reached.

5.  Ideally, gradient descent converges to a local minimum of the cost function, providing the optimal parameters for the model.


Gradient descent is a powerful and widely-used optimization technique in machine learning that allows for efficient training of models. By iteratively updating parameters in the direction of the steepest descent of the cost function, it finds optimal solutions for a variety of applications, from linear regression to deep learning. Understanding how gradient descent works is crucial for effectively applying it in real-world scenarios.

````python
import numpy as np
import pandas as pd

class LogisticRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.intercept = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z)) # this is the actual sigmoid function, what we are doing in this model is finding the best parameters to plug into this.

    def fit(self, X, y):

        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.intercept = 0


        for _ in range(self.iterations): # Think of this loop and the iterations as the legnth of time to which the model fits the data
            linear_model = np.dot(X, self.weights) + self.intercept # this just helps us get the linear model
            predictions = self.sigmoid(linear_model) # plugs in the linear model into the sigmoid function to make


            # Obtaining the gradients
            dw = (1 / n_samples) * np.dot(X.T, (predictions - y))
            db = (1 / n_samples) * np.sum(predictions - y)

            self.weights -= self.learning_rate * dw # Based off of the gradient, decrement value [learning_rate] from the weights
            self.intercept -= self.learning_rate * db * dw # Based off of the gradient, decrement value [learning_rate] from the intercept


````

For the reason of which the gradients calculated are formulated in that way, it is done by taking the gradient of the associated loss function.

As the Logistic reggression model classifies between two options, it's loss function(a mathematical function that quantifies the difference between the predicted values and actual values) is of binary cross entropy.

Now, for the utilization of the model, lets craft a quick prediction function

````python
import numpy as np
import pandas as pd

class LogisticRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.intercept = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):

        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.intercept = 0

        for _ in range(self.iterations):
            linear_model = np.dot(X, self.weights) + self.intercept
            predictions = self.sigmoid(linear_model)


            dw = (1 / n_samples) * np.dot(X.T, (predictions - y))
            db = (1 / n_samples) * np.sum(predictions - y)

            self.weights -= self.learning_rate * dw
            self.intercept -= self.learning_rate * db

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.intercept
        predictions = self.sigmoid(linear_model)
        return [1 if i > 0.5 else 0 for i in predictions]
````
