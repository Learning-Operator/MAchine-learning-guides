## Supervised learning

Supervised learning is a fundamental technique in machine learning where an algorithm learns from a labeled dataset to make predictions or decisions. In this approach, the training data consists of input-output pairs, where each input is associated with a correct label or target. The algorithm's goal is to learn the mapping from inputs to outputs by minimizing the difference between its predictions and the actual labels. Once trained, the model can generalize this knowledge to make accurate predictions on new, unseen data. Supervised learning is widely used in applications such as image classification, medical diagnosis, and stock price prediction.

```python
# Scikit-learn Example: Simple Machine Learning Model
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a RandomForest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Predict and calculate accuracy
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of RandomForest Classifier: {accuracy}")
```

The Random forest classifier is an extremely popular model of which is a supervised Machine learning algorithm.

The random forest classifier is such because of how it:

1. Is trained on Labeled data
2. Its goal is to explicitely map a functional relationship between the input and corresponding output data.
3. Uses a loss function to measure how well its predictions perform. Comparing labelled outputs.




### Reggresion

Regression is a statistical method used to understand and quantify the relationship between one or more independent variables and a dependent variable. Its primary purpose is to predict continuous outcomes, such as estimating a house's price based on features like size, location, and number of rooms. In machine learning, regression algorithms, such as linear regression, polynomial regression, and ridge regression, are trained on data to establish the relationship between input and output variables. The model then uses this relationship to make predictions on new data points. Regression analysis also helps identify trends, relationships, and correlations within the data, which can be useful for forecasting, risk assessment, and various decision-making processes.

### Classification

Classification is a machine learning technique focused on categorizing data into distinct classes or labels. Unlike regression, where the output is continuous, classification predicts discrete classes, such as determining if an email is spam or not or identifying the species of an animal from a photo. Popular classification algorithms include decision trees, support vector machines, and logistic regression. In a classification problem, the model learns patterns in labeled training data, which helps it assign class labels to new, unseen data. Classification is widely used in applications like image and speech recognition, medical diagnosis, and fraud detection, where sorting information into accurate categories is crucial.

### Neural Networks

Neural networks are a class of machine learning models inspired by the human brain's structure, composed of interconnected nodes called neurons. These networks consist of multiple layers – an input layer, one or more hidden layers, and an output layer – each containing a series of neurons that process data. The layers work together to learn complex patterns by adjusting connections (weights) during training. Neural networks are highly flexible and can tackle tasks ranging from image recognition to natural language processing. With various architectures like convolutional neural networks (CNNs) and recurrent neural networks (RNNs), they excel at identifying patterns in large datasets, making them foundational in deep learning and powering innovations in artificial intelligence.

## Unsupervised learning

Unsupervised learning is a type of machine learning where the algorithm is trained on data without labeled outputs. Instead of being told what the correct answers are, the algorithm tries to discover hidden patterns or structures within the data. Common tasks in unsupervised learning include clustering (grouping similar data points) and dimensionality reduction (simplifying data while preserving its important characteristics). Since there are no explicit labels or targets, unsupervised learning is useful for exploring data, identifying patterns, and finding insights in unstructured datasets, such as customer segmentation or anomaly detection.








```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs # just a neat method to create clusters of data
from sklearn.cluster import KMeans

# Generate sample data with 3 clusters
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42) #make_blobs returns an array of genarated samples as well as its shape, in this case we dont want that

# Plot the sample data
plt.scatter(X[:, 0], X[:, 1], s=30, color='gray') # We are just plotting the data
plt.title("Sample Data for Clustering") # We are just plotting the data
plt.show() # We are just plotting the data

# Define the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42) #  This is the unsupervised learning model of whcih finds the clusters to group the data, we are initializing its class

# Fit the model to the data
kmeans.fit(X)

# Predict cluster labels for each data point
labels = kmeans.predict(X) # just its predictions, because we are looking at clustering, its labels are coordinate points

# Plot the clustered data with centroids
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30) # We are just plotting the data
centroids = kmeans.cluster_centers_# We are just plotting the data
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200) # We are just plotting the data
plt.title("Data Clustered using KMeans")# We are just plotting the data
plt.show()# We are just plotting the data
```

The code above illustrates a unsupervised learning model find clusters.
